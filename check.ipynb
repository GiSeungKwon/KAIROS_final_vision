```python
import torch

# 1. CUDA를 사용할 수 있는지 확인 (True/False)
is_cuda_available = torch.cuda.is_available()
print(f"CUDA 사용 가능 여부: {is_cuda_available}")

if is_cuda_available:
    # 2. 시스템에 설치된 GPU 개수 확인
    gpu_count = torch.cuda.device_count()
    print(f"사용 가능한 GPU 개수: {gpu_count}")

    # 3. 현재 사용 중인 GPU의 이름 확인
    current_device_index = torch.cuda.current_device()
    gpu_name = torch.cuda.get_device_name(current_device_index)
    print(f"현재 GPU 장치 이름: {gpu_name}")

    # 4. PyTorch가 빌드된 CUDA 버전 확인
    cuda_version = torch.version.cuda
    print(f"PyTorch 빌드 CUDA 버전: {cuda_version}")
else:
    print("CUDA 장치를 찾을 수 없습니다. GPU 드라이버 및 PyTorch 설치를 확인하세요.")
```
